```python
# BST constructor

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None
        
    def insert(self, value):
        new_node = Node(value)
        if self.root is None:
            self.root = new_node
            return True
        
        temp = self.root
        
        while True:
            if new_node.value == temp.value:
                return False
            if new_node.value < temp.value:
                if temp.left is None:
                    temp.left = new_node
                    return True
                temp = temp.left
            else:
                if temp.right is None:
                    temp.right = new_node
                    return True
                temp = temp.right
                
    def lookup(self, value):
        temp = self.root
        
        while temp is not None:
            if value < temp.value:
                temp = temp.left
            elif value > temp.value:
                temp = temp.right
            else:
                return True
        
        return False
    
    def BFS(self, root):
        current_node = self.root
        my_list = []
        myqueue = []
        height = 0
        myqueue.append(current_node)
        
        while len(myqueue) > 0:
            current_node = myqueue.pop(0)
            my_list.append(current_node.value)
            
            if current_node.left:
                myqueue.append(current_node.left)
            if current_node.right:
                myqueue.append(current_node.right)
        
        return my_list, height
    
    def BFSR(self, queue, rlist):
        if len(queue) == 0:
            return rlist
        current_node = queue.pop(0)
        rlist.append(current_node.value)

        if current_node.left:
            queue.append(current_node.left)
        if current_node.right:
            queue.append(current_node.right)
            
        return self.BFSR(queue, rlist)
        
            
        
    
    
```
